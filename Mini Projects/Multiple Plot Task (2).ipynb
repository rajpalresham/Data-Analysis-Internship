# Multiple plot task


```python
import pandas as pd
unrate = pd.read_csv('unrate.csv')
unrate
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate
```

```python
twelve_values = unrate.head(12)
twelve_values
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_values = twelve_values['DATE']
y_values = twelve_values['VALUE']
```

```python
plt.figure(figsize=(6,4))
plt.plot(x_values,y_values)
plt.xticks(rotation = 90)
plt.xlabel("Months")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])

plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(5,1,1)
ax1.plot(unrate['DATE'][0:12], unrate['VALUE'][0:12],c="Brown") 
ax1.set_title("Monthly Employment Trends,1948")
ax1.set_xlabel("Month")
ax1.set_ylabel("Unemployment Rate")

ax2=fig.add_subplot(5,1,2)
ax2.plot(unrate['DATE'][12:24], unrate['VALUE'][12:24],c="green") 
ax2.set_title("Monthly Employment Trends,1949")
ax2.set_xlabel("Month")
ax2.set_ylabel("Unemployment Rate")

ax3=fig.add_subplot(5,1,3)
ax3.plot(unrate['DATE'][24:36],unrate['VALUE'][24:36],c='Orange')
ax3.set_title("Monthly Employment Trends,1950")
ax3.set_xlabel("Month")
ax3.set_ylabel("Unemployment Rate")

ax4=fig.add_subplot(5,1,4)
ax4.plot(unrate['DATE'][36:48], unrate['VALUE'][36:48],c="red")
ax4.set_title("Monthly Employment Trends,1951")
ax4.set_xlabel("Month")
ax4.set_ylabel("Unemployment Rate")

ax4=fig.add_subplot(5,1,5)
ax4.plot(unrate['DATE'][48:60], unrate['VALUE'][48:60],c="blue")
ax4.set_title("Monthly Employment Trends,1951")
ax4.set_xlabel("Month")
ax4.set_ylabel("Unemployment Rate")
fig.tight_layout()

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red', label='1948')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='black', label='1949')
plt.plot(unrate[24:36]['MONTH'], unrate[24:36]['VALUE'], c='blue', label='1950')
plt.plot(unrate[36:48]['MONTH'], unrate[36:48]['VALUE'], c='green', label='1951')
plt.plot(unrate[48:60]['MONTH'], unrate[48:60]['VALUE'], c='orange', label='1952')
plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month,Integer")
plt.ylabel("Unemployment Rate, Percent")
```
