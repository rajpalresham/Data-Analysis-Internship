# 1. Bar Plot

```python
import pandas as pd
reviews = pd.read_csv('fandango_scores.csv')
reviews
```

```python
norm_reviews=reviews[['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Stars','Fandango_Ratingvalue']]
norm_reviews
```

```python
norm_reviews.head(1)
```

```python
import matplotlib.pyplot as plt
fig, ax=plt.subplots()
## positions of the left side of the 5 bars [0.75, 1.75, 2.75, 3.75, 4.75]
import numpy as np
bar_positions = np.arange(5) + 0.75

## Heights of the bars, In our cases, the avergae rating for the first movie in the dataset.
num_columns = ['RT_user_norm',"Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
bar_heights = norm_reviews[num_columns].iloc[0].values
ax.bar(bar_positions, bar_heights, 0.5)
tick_positions = range(1,6)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers: Age of Ultron")
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_columns, rotation=75)
plt.show()
```

```python

num_columns = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
fig, ax = plt.subplots()
bar_widths = norm_reviews[num_columns].iloc[0].values
bar_positions = np.arange(5) + 0.75
ax.barh(bar_positions, bar_widths, 0.5)
ytick_positions = range(1,6)
ax.set_yticks(tick_positions)
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
ax.set_yticklabels(num_cols)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.barplot(x=norm_reviews.index, y=norm_reviews['IMDB_norm'])
```

# 2. Scatter Plot

```python
fig, ax = plt.subplots()
ax.scatter(norm_reviews["Fandango_Ratingvalue"], norm_reviews["RT_user_norm"])
ax.set_xlabel("Fandango")
ax.set_ylabel("Rotten Tomatoes")
plt.show()
```

```python
fig = plt.figure(figsize=(5,10))
fig, ax = plt.subplots()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.scatter(norm_reviews["Fandango_Ratingvalue"].values, norm_reviews['RT_user_norm'].values,c="orange")
ax1.set_xlabel('Fandango')
ax1.set_ylabel("Rotten Tomatoes")
ax2.scatter(norm_reviews["RT_user_norm"].values, norm_reviews["Fandango_Ratingvalue"].values,c="red")
ax2.set_xlabel("Rotten Tomatoes")
ax2.set_ylabel("Fandango")
plt.show()
```
