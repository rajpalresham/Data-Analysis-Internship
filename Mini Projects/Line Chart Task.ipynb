```python
a = twelve_values['DATE']
b = twelve_values['VALUE']
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.lineplot(a,b)
plt.xticks(rotation=40)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```
