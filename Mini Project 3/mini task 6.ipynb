```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv('cardioActivities.csv',parse_dates=True,index_col="Date")
df_activities.head(2)
```

```python
df_run=df_activities.loc[df_activities['Type'] == 'Running'].copy()
runs_subset_2015_2018 = df_run.loc["2019-01-01 14:05:12":"2015-01-01 18:35:11"]
```

```python
runs_distance = runs_subset_2015_2018["Distance (km)"]
runs_hr = runs_subset_2015_2018["Average Heart Rate (bpm)"]
```

```python
fig, (ax1,ax2) = plt.subplots(2, sharex=True, figsize=(12,8))
runs_distance.plot(ax=ax1 , color="Blue")  #first plot
ax1.set(ylabel="Dsitance (km)", title="Avergae Data")
ax1.axhline(runs_distance.mean(),color="black",linewidth = 1, linestyle = '-.')

runs_hr.plot(ax=ax2, color="Brown")
ax2.set(xlabel="Date", ylabel="Average Heart Rate (bpm)")
ax2.axhline(runs_hr.mean(), color="black",linewidth = 1, linestyle = '-.')

plt.show()
```
