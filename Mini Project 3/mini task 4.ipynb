```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv('cardioActivities.csv',parse_dates=True,index_col="Date")
df_activities.head(5)
```

```python
df_run=df_activities.loc[df_activities['Type'] == 'Running'].copy()
runs_subset_2013_2018 = df_run.loc["2018-11-11 14:05:12":"2013-01-02 18:35:11"] #subset from 13-18
runs_subset_2013_2018.plot(subplots=True,figsize=(6,4), marker="*", markersize=8)
plt.show()
```
