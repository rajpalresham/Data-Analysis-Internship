```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv('cardioActivities.csv',parse_dates=True,index_col="Date")
df_activities.head(2)
```

```python
import statsmodels.api as sm
runs_subset_2013_2018 = df_activities.loc["2018-01-01 14:05:12":"2013-01-01 18:35:11"]
df_run_dist_wkly =  runs_subset_2013_2018["Distance (km)"].resample("W").bfill()
```

```python
fig = plt.figure(figsize = (12,5))
ax = df_run_dist_wkly.plot(marker="*",label = "Trend", color="Brown")
ax.legend()
ax.set_title = 'Running Title'
plt.show()
```
